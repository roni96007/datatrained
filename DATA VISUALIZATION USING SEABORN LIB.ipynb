# SEABORN VISUALIZATION

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df1=pd.read_csv('empl.csv')
df1
```

STRIP PLOT()

```python
x=['sunday','monday','tuesday','wednesday','thrusday','friday','saturday']
y=[2,4,6,8,10,12,14]
sns.stripplot(x,y)
```

```python
sns.stripplot(df1['Age'])
plt.show()
```

Bar plot()

```python
a=['sunday','monday','tuesday','wednesday','thrusday','friday','saturday']
b=[2,4,6,8,10,12,14]
sns.barplot(a,b)
```

```python
sns.barplot(data=df1)
```

Violin Plot

```python
x=['sunday','monday','tuesday','wednesday','thrusday','friday','saturday']
y=[2,4,6,8,10,12,14]
sns.violinplot(x,y)
```

```python
#create violinplot of Salary column
#1st method
sns.violinplot(df1['Salary'])   #using (dataframe[column Name])
#show the plot
plt.show()
```

```python
#Different method
sns.violinplot(x='Age', data=df1)  #using (x=column name , data=dataframe)
#show the plot
plt.show()
```

Count Plot()

```python
sns.countplot(df1['Age'])
```
