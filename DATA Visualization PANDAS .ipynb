# PANDAS DATA VISUALIZATION

```python
import pandas as pd
import numpy as np
```

bar plot

```python
df1=pd.DataFrame(np.random.rand(10,5),columns=['a','b','c','d','e'])
print(df1)
df1.plot.bar()
df1.plot.barh()
```

```python
df1.plot.bar(stacked=True)
df1.plot.barh(stacked=True)
```

```python
df2=pd.DataFrame(np.random.randn(7,4),columns=[1,2,3,4])
print(df2)
df2.plot.bar()
df2.plot.barh()
df2.plot.bar(stacked=True)
df2.plot.barh(stacked=True)
```

```python
df3=pd.DataFrame(np.random.rand(8,5),columns=['A','B','C','D','E'])
print(df3)
df3.plot.bar()
df3.plot.barh()
df3.plot.bar(stacked=True)
df3.plot.barh(stacked=True)
```

# CHANGE THE INDEX

how to change the index values in dataframe. Create a Dataframe with some key value pairs in a dictionary and change the index values.

```python
df=pd.DataFrame({'Day':[1,2,3,4,5],'visitors':[100,200,300,400,500],'bouncer':[200,400,600,800,1000]})
print(df)
df4=df.set_index('Day')
df4
```

```python
df5=pd.DataFrame({'Day':[1,2,3,4,5,6,7,8],'visitors':[100,200,300,400,500,600,700,800],'bouncer':[200,400,600,800,1000,1200,1400,1600]})
df5.set_index('Day',inplace=True)
dframe=df5.rename(columns={"visitors":"Users"})
print(dframe)
```

```python
df6=df.rename(columns={'bouncer':'users'})
print(df6)
```
